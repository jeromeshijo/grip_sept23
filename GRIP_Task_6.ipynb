```python
from six import StringIO
from IPython.display import Image
from sklearn.tree import export_graphviz
import pydotplus
```

```python
iris = datasets.load_iris()
dataset = pd.DataFrame(iris.data, columns = iris.feature_names)
dataset.head()
```

###**3) Visualizing The Decision Tree**

```python
X = dataset.iloc[:,:].values
y = iris.target
decision_tree = DecisionTreeClassifier()
# decision_tree.fit()
```

```python
X[0:5]  # head of dataset
```

```python
y  # values of dataset
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y)
decision_tree.fit(X_train,y_train)
```

```python
dot_data = StringIO()
export_graphviz(decision_tree, out_file=dot_data, feature_names=iris.feature_names,
                filled=True, rounded=True,
                special_characters=True,class_names=["Iris-Setosa","Iris-Versicolor","Iris-Virginica"])
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())
Image(graph.create_png())
```

###**4) Testing the Algorithm**

```python
y_pred = decision_tree.predict(X_test)
```

```python
y_test
```

```python
y_pred
```

```python
accurate=0
for i in range(len(y_test)):
  if (y_test[i]==y_pred[i]):
    accurate+=1
print("Accuracy of Testing Data: ",(accurate/i)*100,"%")
```
